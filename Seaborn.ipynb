```python
#Histogram
plt.figure(figsize=(10,5))
sns.histplot(df['mpg'], bins=20);
#sns.histplot(df.mpg)
plt.title('MPG Distribution');
```

```python
#Histogram
plt.figure(figsize=(10,5))
sns.histplot(df, x='mpg', hue="origin", multiple="stack", alpha = 0.7);
#sns.histplot(df.mpg)
plt.title('MPG Distribution');
```

```python
#JoinPlot
sns.jointplot(x="mpg", y="weight", data=df, kind="scatter");
```

```python
sns.jointplot(x="mpg", y="weight", data=df, kind="hex");
sns.jointplot(x="mpg", y="weight", data=df, kind="reg");
sns.jointplot(x="mpg", y="weight", data=df, kind="kde");
```

```python
sns.pairplot(df, hue="origin",palette="coolwarm");
```

```python
sns.pairplot(df,palette="coolwarm");
```

```python
sns.kdeplot(data=df, x='mpg', hue="origin");
plt.title("KDE chart")
```

```python
# BarPlot
sns.barplot(x="origin", y="horsepower", data=df);
```

```python
sns.barplot(x="origin", y="horsepower", data=df, estimator=np.std);
```

```python
# CountPlot
sns.countplot(x="origin", data=df);
```

```python
df['mpg_level']=df['mpg'].apply(lambda x: 'Low' if x<15 else 'High' if x>30 else "Medium")
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="model_year", hue="mpg_level", data= df);
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="model_year", hue="cylinders", data= df);
```

```python
#RelPlot
sns.relplot(x='model_year', y='mpg', data=df);
```

```python
sns.relplot(x='horsepower', y='mpg', hue='origin', data=df);
```

```python
#CatPlot
sns.catplot(data=df, kind="bar", x='origin', y='horsepower', hue='cylinders', alpha=0.7, height=4);
```

```python
#SwarmPlot
sns.swarmplot(x='origin', y='mpg', data=df);
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='origin', y='mpg', hue='cylinders', data=df);
```
