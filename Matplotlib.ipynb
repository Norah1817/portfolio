```python
Moves_count = pd.DataFrame(df.release_year.value_counts(ascending=True))
Moves_count
#عدد الأفلام لكل سنة
```

```python
Moves_count = Moves_count.reset_index()
Moves_count
#   و أستخدمنا رست اندكس عشان يسوي عمود اندكس عشان ما يعتبر ريلايس يير اندكس 
```

```python
Moves_count.columns = ['year', 'count'] # هنا رتبنا أسماء الأعمدة 
Moves_count = Moves_count.sort_values(by=['year']) # هنا رتبنا الاعمدة بالسنوات من الأصغر للأكبر
Moves_count
# الآن جاهزة نرسمها باللين شارت 
```

```python
plt.plot(Moves_count['year'], Moves_count['count']);
# و بكذا جاوبنا على سؤال كم عدد الأفلام التي يتم انتاجها كل سنة 
```

```python
# تغيير خصائص الرسمة 
plt.plot(Moves_count['year'], Moves_count['count'], c= '0.2'); # كتبنا رقم اللون على شكل سترنق 
```

```python
plt.plot(Moves_count['year'], Moves_count['count'], c= '0.2');
plt.xlabel('Year')
plt.ylabel('Movies count ')
plt.title("Number of Released Movies Annually");
plt.xlim(1960, 2015)# خاصية تسمح لنا نختار حدود المحاور و تختفي الفراغات 
plt.xticks(range(1960, 2016, 5));
#  و هنا نقسم المدة بين السنوات 
# plt.axis('off'); # بهذي الخاصية نخفي محاور ال x و y 
```

```python
# رسم ثاني يقارنها بعدد المسلسلات 
plt.plot(Moves_count['year'], Moves_count['count'], c= 'red',marker = 'o',markersize = 10, markerfacecolor='b' );
plt.plot(range(1960, 2015, 10), [500, 400,300,200,100, 600], c= 'green', linestyle = ':', linewidth = 4);# انشاء قيمة افتراضية تحدد عدد السنوات 
plt.legend(['Movies', 'Series'], shadow = True, loc=0) # اذا كان الموقع يساوي 0 راح يحدد أفضل موقع لللجند
plt.xlabel('Year')
plt.ylabel('Movies count ')
plt.title("Number of Released Movies Annually");
```

```python
#Bar Chart
# مطلوب نجاوب على سؤال اعلى خمس أفلام من ناحية الأرباح
```

```python
top_revenue = df.groupby('original_title').sum().sort_values(by='revenue', ascending=False)
top_revenue.head() # راح يرجع لنا أعلى خمسة افلام من ناحية الأرباح 
#نستخدم القروب باي لتجميع البيانات على حسب عنوان الفلم و نرتب الفلم على حسب الأرباح نرتبها من الأصغر للأكبر  
```

```python
#و من هذا الجدول نحتاج اسماء الأفلام و مقدار الأرباح نسندهل لمتغير و نرجع اندكس للجدول 
movies = top_revenue.head().index
movies
# النتيجة أسماء الأفلام 
```

```python
# و نسوي متغير كمان للأرباح 
revenue = top_revenue.head().revenue # و هذا عشان ترجع الأرباح باسم الفلم 
revenue
```

```python
revenue = top_revenue.head().revenue.values # وعشان نرجع الأرباح فقط 
revenue
```

```python
# و الآن نعدل أسماء اللأفلام عشان تظهر في الرسمة مرتبة 

#movies = ['Avatar','Star Wars','Titanic','The Avengers','J World']
```

```python
# الآن نقدر نمثل الرسم عندنا متغير لأسماء الأفلام و متغير لللأرباح 
plt.figure(figsize=(12,10))
color = ['blue', '#5B00B2']
plt.bar(movies, revenue, color= color);
plt.xlabel('movies', fontsize=20)
plt.ylabel('revenue', fontsize=20)
#plt.grid()
plt.title('Top 5 movies based on revenue');
```

```python
#Pie chart 
# نجاوب على سؤال ماهي أعلى فئات الأفلام من حيث الربح 
top_genres = df.groupby('genres').sum().sort_values(by='revenue', ascending= False)['revenue'][0:5]
top_genres
# و هنا نحتاج السم لأنها بنجمع الأرباح لكل فئةو سورت عشان التجميع يتم بناءا على القيم الأعلى  
```

```python
plt.pie(top_genres.values, labels= top_genres.index);
```

```python
coors = ['#ff9999', 'b','g','r', '#99ff99']
plt.pie(top_genres.values, labels= top_genres.index, colors= coors, startangle=90, radius=1, autopct='%1.2f%%');
```

```python
plt.pie(top_genres.values, labels= top_genres.index,explode = [0.2,0,0,0.3,0]);
```

```python
#Scatter Chart
# العلاقة بين معدل التصويت والأرباح 
plt.scatter(df.vote_average, df.revenue);
plt.xlabel('vote Average')
plt.ylabel('Revenue')
plt.title('Relationship between vote Average and Revenue')
```

```python
plt.scatter(df.vote_average, df.revenue, alpha=0.7, marker='*',s=150, label='vote Average');
plt.scatter(df.vote_count, df.revenue, alpha=0.8, marker='+', s=150, label='vote count');
plt.legend(loc='upper right')
plt.xlabel('vote Average & count')
plt.ylabel('Revenue')
plt.title('Relationship between vote Average & count vs. Revenue')
```
